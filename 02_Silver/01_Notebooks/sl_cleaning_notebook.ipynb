```python
customers_df = spark.read.table("global_warehouse.raw_global_warehouse.br_customers")
display(customers_df)
```

```python
customers_df = customers_df.na.replace("TEST", "Test", 'customer_type')
customers_df = customers_df.na.replace(["U.S.A", "india"], ["USA","India"], 'country')
display(customers_df)
```

```python
exchange_rates_df = spark.read.table("global_warehouse.raw_global_warehouse.br_exchange_rates")
display(exchange_rates_df)
```

```python
invoice_line_items_df = spark.read.table("global_warehouse.raw_global_warehouse.br_invoice_line_items")
display(invoice_line_items_df)
```

```python
from pyspark.sql.functions import col, greatest, lit
invoice_line_items_df = invoice_line_items_df.na.fill({'discount': 0.0})
invoice_line_items_df = invoice_line_items_df.na.fill({'unit_price': 0})
display(invoice_line_items_df)
```

```python
invoices_df = spark.read.table("global_warehouse.raw_global_warehouse.br_invoices")
display(invoices_df)
```

```python
invoices_df = invoices_df.na.replace(["EAST", "west","North","south"], ["East","West","North","South"], 'region')
invoices_df = invoices_df.na.replace(["paid"], ["Paid"], 'invoice_status')
display(invoices_df)
```

```python
payments_df = spark.read.table("global_warehouse.raw_global_warehouse.br_payments")
display(payments_df)
```

```python
products_df = spark.read.table("global_warehouse.raw_global_warehouse.br_products")
display(products_df)
```

```python
products_df = products_df.na.fill({'cost_price': 0})
products_df = products_df.na.replace(["electronics"], ["Electronics"], 'category')
display(products_df)
```

```python
regions_df = spark.read.table("global_warehouse.raw_global_warehouse.br_regions")
display(regions_df)
```

```python
products_df = products_df.dropDuplicates()
invoices_df = invoices_df.dropDuplicates()
invoice_line_items_df = invoice_line_items_df.dropDuplicates()
customers_df = customers_df.dropDuplicates()
payments_df =payments_df.dropDuplicates()
exchange_rates_df = exchange_rates_df.dropDuplicates()
display(customers_df )
```

```python
catalog_name = "global_warehouse"
schema_name = "staging_global_warehouse"
table_name = "sl_customers"
table_path = f"{catalog_name}.{schema_name}.{table_name}"

customers_df.write.mode("overwrite").saveAsTable(table_path)
```
